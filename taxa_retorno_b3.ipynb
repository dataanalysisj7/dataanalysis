```python
# Identificação de valores constantes na mesma coluna, por erro ou inatividade que não nos interessa.
constant_columns = []
for col in dt.columns:
    if dt[col].nunique() == 1:
        constant_columns.append(col)

if constant_columns:
    print(f"Columns with constant values: {constant_columns}")
else:
    print("No columns with constant values found.")
```

```python
dt.drop(columns=constant_columns, inplace=True)
display(dt.head())
print(f"Remaining columns: {dt.shape[1]}")
```

#### Taxa de retorno simples

```python
dataset_rs = dataset.copy()
```

```python
dataset_normalizado = dataset.copy()
for i in dataset_normalizado.columns[1:]:
  dataset_normalizado[i] = (dataset_normalizado[i] / dataset_normalizado[i][0])
```

```python
dataset_normalizado.drop(labels=['Date'], axis=1, inplace=True)
```

```python
retorno_carteira = (dataset_normalizado / dataset_normalizado.shift(1)) - 1
retorno_carteira.head()
```

```python
retorno = retorno_carteira.mean() * 488 * 100
retorno
```

### Visualização dos Retornos Anuais

Com mais de 200 colunas, plotar todas em um gráfico de barras pode ser difícil de ler. Para uma melhor visualização, vamos focar nos 10 melhores e 10 piores desempenhos anuais. Primeiro, ordenaremos os retornos e selecionaremos esses subconjuntos.

```python
# Ordenar os retornos anuais para identificar os melhores e piores desempenhos
retorno_sorted = retorno.sort_values(ascending=False)

# Selecionar os 10 melhores e os 10 piores
top_10_stocks = retorno_sorted.head(10)
bottom_10_stocks = retorno_sorted.tail(10)

# Concatenar para um único DataFrame para plotagem
selected_stocks = pd.concat([top_10_stocks, bottom_10_stocks])

display(selected_stocks)
```

Agora, vamos gerar um gráfico de barras para visualizar esses retornos  selecionados.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))
# Use the 'retorno_anual' for the plot, which is now sorted and contains top/bottom 10
selected_stocks.plot(kind='bar', color=['green' if x > 0 else 'red' for x in selected_stocks.values])
plt.title('Top 10 e Bottom 10 Retornos Anuais de Ações da B3 (em %)')
plt.xlabel('Ativo (Ticker)')
plt.ylabel('Retorno Anual (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
